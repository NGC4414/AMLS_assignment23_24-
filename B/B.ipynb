```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

def load_and_preprocess_pathmnist():
    pathmnist_path = r"C:\Users\clara\Documents\uni\4thYear_AMLS\Datasets\pathmnist.npz"
    pathmnist= np.load(pathmnist_path)

    # Load the dataset
    with np.load(pathmnist_path, allow_pickle=True) as data:
        x_train = data['train_images']
        y_train = data['train_labels']
        x_val = data['val_images']
        y_val = data['val_labels']
        x_test = data['test_images']
        y_test = data['test_labels']

    # Normalize the images to have a mean of 0.5 and std of 0.5
    train_datagen = ImageDataGenerator(rescale=1./255)
    val_datagen = ImageDataGenerator(rescale=1./255)
    test_datagen = ImageDataGenerator(rescale=1./255)

    # Create the data generators
    train_generator = train_datagen.flow(x_train, y_train)
    val_generator = val_datagen.flow(x_val, y_val)
    test_generator = test_datagen.flow(x_test, y_test)

    return train_generator, val_generator, test_generator

train_gen, val_gen, test_gen = load_and_preprocess_pathmnist()
```

```python
from keras.preprocessing.image import ImageDataGenerator
pathmnist_path = r"C:\Users\clara\Documents\uni\4thYear_AMLS\Datasets\pathmnist.npz"

# Define data directories
train_data_dir = pathmnist_path
test_data_dir = pathmnist_path

# Define data generators with augmentation and preprocessing
batch_size = 32
image_size = (28, 28)
data_generator = ImageDataGenerator(
    rescale=1.0 / 255.0,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Load and preprocess training data
train_generator = data_generator.flow_from_directory(
    train_data_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    seed=42
)

# Load and preprocess test data
test_generator = data_generator.flow_from_directory(
    test_data_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False,
    seed=42
)

# Display dataset information
print("Training Dataset:")
print(train_generator.class_indices)  # Class names and corresponding labels
print(f"Number of training samples: {train_generator.samples}")
print("===================")
print("Test Dataset:")
print(test_generator.class_indices)  # Class names and corresponding labels
print(f"Number of test samples: {test_generator.samples}")
```
